```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import pandas as pd
import scipy.optimize as opt
import scipy.fftpack as fft
from scipy import stats, interpolate, signal
from pylab import plot, show, axis
from numpy.linalg import norm
import cvxpy as cvx
import random
plt.rcParams['figure.figsize'] = [12, 8]
```

Zadanie1 1/2

```python
n = 1000
t = np.linspace(-30, 30, n)
y = 3 * t**2

for i in range(len(y)):
  if i%10 == 0:
    y[i] += np.random.uniform(-100,100)

plt.figure(figsize=(10,15))
plt.subplot(411)
plt.plot(t,y)


m = 200
pr = np.random.choice(n, m, replace=False)
pr.sort()
t2 = t[pr]
b = y[pr]

A = fft.idct(np.identity(n), norm='ortho', axis=0)
A = A[pr]

vx = cvx.Variable(n)
objective = cvx.Minimize(cvx.norm(vx, 1))
constraints = [A*vx == b]
prob = cvx.Problem(objective, constraints)
result = prob.solve(verbose=True)

x = np.array(vx.value)
x = np.squeeze(x)
f = fft.idct(x, norm='ortho', axis=0)

plt.subplot(412)
plt.plot(t, f)
```

```python
n = 100
t = np.linspace(0, 5*np.pi, n)
y = np.sin(t)

def wyliczZera(y):
  mz = []
  for x in range(len(y)-1):
    if (y[x] > 0 and y[x+1] < 0) or (y[x] < 0 and y[x+1] > 0):
      print('dodaje_1')
      mz.append((t[x] + t[x+1])/2)
    if y[x] == 0:
      print('dodaje_2')
      mz.append(t[x])
  if y[x] == 0:
    print('dodaje_3')
    mz.append(t[x])
  print(mz)

wyliczZera(y)
plt.plot(t, y, t, y*0)
```

```python
n = 100
t = np.linspace(0, 1/8, n)
y = np.sin(13 * np.pi * t) + np.sin(46 * np.pi * t) + np.sin(86 * np.pi * t)

gyy = 0
yavr = np.average(y)
m = 5
for k in range(m, len(y)):        # zaczynam od m poniewaz indeksy nie beda sie zgadzac w przeciwnym wypadku
    gyy += (y[k] - yavr)*(y[k-m] - yavr)

gyy = gyy/(len(y)-1)
print(gyy)

kvalue = 0
ystd = np.std(y)
for k in range(1, len(y)):
    kvalue += (y[k] - yavr)**2

kvalue = kvalue/(len(y) * ystd**4)
print(kvalue)

emg = y.copy()
for i in range(0, len(emg)-1):
    emg[i] = y[i]**2 - y[i-1] * y[i+1]

print(emg)

emg = pd.Series(emg)
print(len(emg))

plt.figure(figsize=(14, 6), dpi=80)
plt.subplot(231)
plt.plot(t, y)
plt.legend(['zwykle'])
plt.subplot(232)
plt.plot(t, emg)
plt.legend(['emg'])
plt.subplot(233)
emg.rolling(window=3).mean().fillna(method='backfill').plot()
plt.legend(['moving average'])
plt.show()
```
