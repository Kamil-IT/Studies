## 2. W tym zadaniu będziemy estymować parametry pewnej prostej więc żeby zasymulować, że nie znamy dokładnie tej prostej to dodamy sobie do niej trochę liczb losowych

```python
x = np.sort(np.random.uniform(0,10,15))
y = 3 + 0.25 * x + 0.1 * np.random.randn(len(x))
y[3] +=6
y[8]-=9
y[13]-=6

def l1_fit(x0,x,y):
    return np.sum(np.abs(x0[0] * x + x0[1] - y))

def l2_fit(x0,x,y):
    return np.sum(np.power(x0[0] * x + x0[1] - y,2))

def l0_fit(x0,x,y):
    return np.sum((np.abs(x0[0] * x + x0[1] - y)) > 0)

def l0_fit_norm(x0,x,y):
    return np.linalg.norm((x0[0] * x + x0[1] - y), ord=0)

def l1_fit_norm(x0,x,y):
    return np.linalg.norm((x0[0] * x + x0[1] - y), ord=1)

def l2_fit_norm(x0,x,y):
    return np.linalg.norm((x0[0] * x + x0[1] - y), ord=2)

xopt1 = opt.fmin(func=l1_fit, x0=[1,1], args = (x,y))
xopt2 = opt.fmin(func=l2_fit, x0=[1,1], args = (x,y))
xopt0 = opt.fmin(func=l0_fit, x0=[1,1], args = (x,y))
xopt0norm = opt.fmin(func=l0_fit_norm, x0=[1,1], args = (x,y))
xopt1norm = opt.fmin(func=l1_fit_norm, x0=[1,1], args = (x,y))
xopt2norm = opt.fmin(func=l2_fit_norm, x0=[1,1], args = (x,y))

y_est1 = xopt1[1] + xopt1[0] * x
y_est2 = xopt2[1] + xopt2[0] * x
y_est0 = xopt0[1] + xopt0[0] * x
y_est0norm = xopt0norm[1] + xopt0norm[0] * x
y_est1norm = xopt1norm[1] + xopt1norm[0] * x
y_est2norm = xopt2norm[1] + xopt2norm[0] * x


plt.plot(x,y,'*', label = 'Points')
plt.plot(x,y_est0,'#009933', label = 'l0 norm')
plt.plot(x,y_est0norm,'#003311', ls='--', label = 'l0 norm_np')
plt.plot(x,y_est1,'#6666ff', label = 'l1 norm')
plt.plot(x,y_est1norm,'#0000ff', ls='--', label = 'l1 norm_np')
plt.plot(x,y_est2,'#ff0000', label = 'l2 norm')
plt.plot(x,y_est2norm,'#800000', ls='--', label = 'l2 norm_np')
plt.ylim([-5,16])
plt.legend()
```

## 3. Będziemy mieli sygnał, który jest złożeniem dwóch sinusów, w kolejnych krokach najpierw wybierzemy sobie 10 procent próbek i na postawie tych 10 procent postaramy się ten sygnał odtworzyć.

```python
n = 5000
t = np.linspace(0, 1/8, n)
y = np.sin(100 * np.pi * t) + np.sin(333 * np.pi * t)

plt.figure(figsize=(10,15))
plt.subplot(411)
plt.plot(t,y)

m = 500 # - 10% probek
pr = np.random.choice(n, m, replace=False) # Losowanie próbek
pr.sort() # sortowanie do wyswietlania
t2 = t[pr]
b = y[pr]

plt.subplot(412)
plt.plot(t2,b)
plt.plot(t2,b, 'r*')

A = fft.idct(np.identity(n), norm='ortho', axis=0)
A = A[pr]

vx = cvx.Variable(n)
objective = cvx.Minimize(cvx.norm(vx, 1))
constraints = [A*vx == b]
prob = cvx.Problem(objective, constraints)
result = prob.solve(verbose=True)

x = np.array(vx.value)
x = np.squeeze(x)
f = fft.idct(x, norm='ortho', axis=0)

plt.subplot(413)
plt.plot(f)
```

## 8.  przygotować kod, który dla funkcji sinus, trójkąt, fala prostokątknej i losowej wyznaczy periodogramy i widma metodą Welcha

```python
# autokorelacja
def workToDo(n, R, t, y):
    kor = []
    for r in range(R):
        result = 0
        for i in range(1, n - r):
            result += y[i] * y[i-r]
        result = result * (1 / (n - r))
        kor.append(result)
    return kor

# def workToDo(signal,m):
#   N =  len(signal)
#   return [(1/(N-r)*sum([signal[i]*signal[i-r] for i in range(N-r)])) for r in range(m)]
```

### SINUS

```python
n = 1000
R = 500
t = np.linspace(0, 5*np.pi, n)
y = np.sin(t)

kor = workToDo(n, R, t, y)
korNp = np.correlate(y, y, "full")

f_periodogram, Pxx_den_periodogram = signal.periodogram(y, n)
f_welch, Pxx_den_welch = signal.welch(y, n)

with plt.style.context('seaborn'): 
    plt.figure(figsize=(14, 6), dpi=80)
    plt.subplot(231)
    plt.plot(t,y)
    plt.subplot(232)
    plt.plot(kor)
    plt.subplot(233)
    plt.plot(korNp)
    plt.subplot(234)
    plt.plot(f_periodogram, Pxx_den_periodogram)
    plt.subplot(235)
    plt.plot(f_welch, Pxx_den_welch)
    plt.show()
```

## 9. wygenerować (z palca) szereg czasowy w bibliotece pandas i korzystając z tej biblioteki, wyliczyć: wartość średnią, wartość minimalną, maksymalną oraz odchylenie standardowe.

```python
time = pd.Series([10,12,16,12,40,16,37,28,19,50,38,60,37,194,62,4,18,39,195,182,45,62])

average = time.mean()
min = time.min()
max = time.max()
median = time.median()
# df = pd.DataFrame(time)
# std = df.std(axis = 0, skipna = True)
std = time.std()

print(f'\naverage: {average}\nmin: {min}\nmax: {max}\nmedian: {median}\nstd: {std}')
```

## 10.  znaleźć w szeregu czasowym wartości odstające oraz wyświetlić ich indeksy

```python
odstajace = []
odstajaceIndex = []
for i in range(len(time)):
    if time.median() + 1*time.std() < abs(time[i]):
        odstajace.append(time[i])
        odstajaceIndex.append(i)
print(odstajace)
print(odstajaceIndex)
```

## 11. interpolacja szeregu czasowego- wychwytujemy wartości odstające, zapominamy o ich oryginalnych wartościach i korzystając z biblioteki pandas wstawić odpowiednie wartości

```python
filtTime = time.copy()
for i in range(len(time)):
  if i in odstajaceIndex:
    filtTime[i] = np.nan

inter = filtTime.interpolate(method='cubic')
for x in range(len(filtTime)):
  print(f'time / inter = {filtTime[x]} / {inter[x]}')
```

## 12. zaimplementować bez wykorzystania wbudowanych funckji okno przesuwne, z uwzględnieniem: czy okna na siebie nachodzą i jeżeli tak to w jakim stopniu. Gdzie: w- szerokość okna, p- procent pokrycia

```python
data = pd.Series([10,12,16,17,40,18,37,28,19,50,38,60,37,94,62,4,18,39,95,82])

def oknoPrzesuwne(w, p):
    procent = w * p
    cechyGrupy = []
    minGrupy = []
    
    index = 0
    cechy = []
    for x in range(len(data)):
        try:
            cechy.append(data[index - procent])
        except KeyError:
            cechy.append(None)
        if len(cechy) >= w:
            cechyGrupy.append(cechy)
            minGrupy.append(np.amin([elem for elem in cechy if elem not in [None]]))
            cechy = []
            if procent > 0:
                for y in range(int(procent)):
                    try:
                        cechy.append(cechyGrupy[-1][-1 - y])
                    except KeyError:
                        cechy.append(None)           
        index += 1
    
    for elem in range(int(procent)):
        cechy.append(data[::-1][elem])
            
    if len(cechy) > 0:
        while len(cechy) < 4:
            cechy.append(None)
        cechyGrupy.append(cechy)
        minGrupy.append(np.amin([elem for elem in cechy if elem not in [None]]))
        cechy = []
        
    return cechyGrupy, minGrupy
```

```python
w = 4  # szerokosc okna przesuwnego
p = 0.5  # procent pokrycia

assert p < 1, "Wartosc nie moze byc wieksza niz 1"

oknoPrzesuwne(w, p)
```

### Tiger

```python
import numpy as np
import matplotlib.pyplot as plt

# simulate EMG signal
burst1 = np.random.uniform(-1, 1, size=50) + 0.08
burst2 = np.random.uniform(-1, 1, size=150) + 0.08
quiet = np.random.uniform(-0.05, 0.05, size=200) + 0.08
emg = np.concatenate([quiet, burst1, quiet, burst2, quiet])
time = np.array([i/1000 for i in range(0, len(emg), 1)]) # sampling rate 1000 Hz

plt.plot(time, emg)
plt.xlabel('Time (sec)')
plt.ylabel('EMG (a.u.)')
```

```python
emgf = emg.copy()
for i in range(0, len(emgf)-1):                       # nie wiem czy (1, len(emgf)-2), czy tak jak jest
    emgf[i] = emg[i]**2 - emg[i-1] * emg[i+1]

plt.plot(time, emgf)
plt.xlabel('Time (sec)')
plt.ylabel('EMG (a.u.)')
```
