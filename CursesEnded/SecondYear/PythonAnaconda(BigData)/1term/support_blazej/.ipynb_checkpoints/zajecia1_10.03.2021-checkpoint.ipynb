```python
import matplotlib.pyplot as plt
from numpy import linspace, cos, pi, ceil, floor, arange, sin, sinc
from pylab import plot, show, axis
```

```python
f = 1 # czestotliwosc
fs = 20 # n probek w jednostce czasu

t = arange(-1, 1 + 1/fs, 1/fs)

x = sin(2 * pi * f * t)

plot(t, x, 'o')
```

```python
# chcemy odtworzyć sinusa, więc potrzebujemy cos bardziej wyrafinowego

f = 1 # czestotliwosc
fs = 3 # n probek w jednostce czasu /  czestotliwość próbkowania

t = linspace(-1, 1, 100)
ts = arange(-1, 1 + 1/fs, 1/fs)

wsp = len(ts) # aktualna wielkość wektora
s_rec = 0

for k in range(-wsp, wsp):
    s_rec = s_rec + sin(2*pi*(k/fs)) * sinc(k - fs * t)


plot(t, s_rec, '--', t, sin(2*pi*t), 'r', ts, sin(2*pi*ts), 'o')
```

```python
f = 1 # czestotliwosc
fs = 3 # n probek w jednostce czasu /  czestotliwość próbkowania / zeby jakosc byla dobra a jak najmniej probek

t = linspace(-1, 1, 100)
ts = arange(-1, 1 + 1/fs, 1/fs)

wsp = len(ts) # aktualna wielkość wektora
s_rec = 0

for k in range(-wsp, wsp):
    s_rec = s_rec + sin(2*pi*(k/fs)) * sinc(k - fs * t)


plot(t, s_rec, '--', t, sin(2*pi*t), t, abs(sin(2*pi*t) - s_rec), '-', t, s_rec*0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from numpy.linalg import norm

x = np.sort(np.random.uniform(0,10,15))
y = 3 + 0.25 * x + 0.1 * np.random.randn(len(x))
yy = 3 + 0.25 *x
def l2_fit(x0,x,y):
  return np.sum(np.power(x0[0]*x+x0[1] - y,2))

def l1_fit(x0,x,y):
  return np.sum(np.abs(x0[0] * x + x0[1] - y))

def l0_fit(x0,x,y):
  return np.sum((np.abs(x0[0] * x + x0[1] - y)) > 0)

def l0_fit_np(x0,x,y):
    return np.linalg.norm((x0[0] * x + x0[1] - y), ord=0)


xopt2 = opt.fmin(func=l2_fit, x0=[1,1], args = (x,y))
xopt1 = opt.fmin(func=l1_fit, x0=[1,1], args = (x,y))
xopt0 = opt.fmin(func=l0_fit, x0=[1,1], args = (x,y))
xopt0np = opt.fmin(func=l0_fit_np, x0=[1,1], args = (x,y))

y_est0np = xopt0np[1] + xopt0np[0] * x
y_est0 = xopt0[1] + xopt0[0] * x
y_est1 = xopt1[1] + xopt1[0] * x
y_est2 = xopt2[1] + xopt2[0] * x

plt.subplot(211)
plt.plot(x,y,'*')
plt.plot(x,y_est1,'g-', label = 'l1 norm')
plt.plot(x,y_est2,'r-', label = 'l2 norm')
plt.plot(x,y_est0,'b-', label = 'l0 norm')
plt.plot(x,y_est0np,'p-', label = 'l0 norm numpy')
plt.plot(x,yy, 'k-', label = 'prosta')
plt.legend()

y2 = y.copy()
y2[3] +=10
y2[13]-=5

xopt22 = opt.fmin(func=l2_fit, x0=[1,1], args = (x,y2))
xopt11 = opt.fmin(func=l1_fit, x0=[1,1], args = (x,y2))
xopt00 = opt.fmin(func=l0_fit, x0=[1,1], args = (x,y))
xopt00np = opt.fmin(func=l0_fit_np, x0=[1,1], args = (x,y))

y_est00np = xopt00np[1] + xopt00np[0] * x
y_est00 = xopt00[1] + xopt00[0] * x
y_est11 = xopt11[1] + xopt11[0] * x
y_est22 = xopt22[1] + xopt22[0] * x

plt.subplot(212)
plt.plot(x,yy,'k', label = 'prosta')
plt.plot(x,y_est11,'g--', label = 'l1 norm')
plt.plot(x,y_est22,'r--', label = 'l2 norm')
plt.plot(x,y_est00,'b--', label = 'l0 norm')
plt.plot(x,y_est00np,'p--', label = 'l0 norm_np')
plt.ylim([-5,16])
plt.legend()
```
