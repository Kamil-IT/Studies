```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import cvxpy as cvx

# Sygnał testowy
n = 5000
t = np.linspace(0, 1/8, n)
y = np.sin(100 * np.pi * t) + np.sin(333 * np.pi * t)

plt.figure(figsize=(10,15))
plt.subplot(411)
plt.plot(t,y)

# Akwizycja danych
m = 500 # 10% próbek
pr = np.random.choice(n, m, replace=False) # losowanie probek
pr.sort() # sortowanie do wyswietlania
t2 = t[pr]
b = y[pr]

plt.subplot(412)
plt.plot(t2,b)
plt.plot(t2,b,'r*')

A = fft.idct(np.identity(n), norm='ortho', axis=0)
A = A[pr]

vx = cvx.Variable(n)
objective = cvx.Minimize(cvx.norm(vx, 1))
constraints = [A*vx == b]
prob = cvx.Problem(objective, constraints)
result = prob.solve(verbose=True)

x = np.array(vx.value)
x = np.squeeze(x)
f = fft.idct(x, norm='ortho', axis=0)

plt.subplot(413)
plt.plot(f)

vx = cvx.Variable(n)
objective = cvx.Minimize(cvx.norm(vx, 2))
constraints = [A*vx == b]
prob = cvx.Problem(objective, constraints)
result = prob.solve(verbose=True)

x = np.array(vx.value)
x = np.squeeze(x)
f_l2 = fft.idct(x, norm='ortho', axis=0)
plt.subplot(414)
plt.plot(f_l2)
```

```python
sqr_err1 = (f-y)**2
sqr_err2 = (f_l2-y)**2
av_sqr_err1 =sum(sqr_err1)/len(sqr_err1)
av_sqr_err2 = sum(sqr_err2)/len(sqr_err2)
plt.subplot(211)
plt.plot(sqr_err1,color='r')
plt.subplot(212)
plt.plot(sqr_err2,color='g')
print(f'average l1 error {av_sqr_err1}')
print(f'average l2 error {av_sqr_err2}')
```

```python
from scipy import interpolate
pr = np.random.choice(n, m, replace=False) # losowanie probek
pr.sort() # sortowanie do wyswietlania
t2 = t[pr]
t2 = np.hstack([[t[0]],t2,[t[-1]]])
b = np.hstack([[y[0]],y[pr],[y[-1]]])
r = interpolate.interp1d(t2, b, kind=3)
ry = r(t)
plt.subplot(311)
plt.plot(ry)
plt.subplot(312)
plt.plot(y)
err =(ry-y)**2
plt.subplot(313)
plt.plot(err)
av_err = sum(err)/len(err)
print(av_err)
```
