```python
import numpy as np
import scipy.signal as scisig
import matplotlib.pyplot as plt


def autocorrelate(signal,m):
  N =  len(signal)
  return [(1/(N-r)*sum([signal[i]*signal[i-r] for i in range(N-r)])) for r in range(m)]
sf = 200
t = np.arange(-5,5,1/sf)
x_sin = np.sin(np.pi*2*t)
t_len =len(t)
m = [i for i in range(int(t_len/2))]
x_sqr = scisig.square(np.pi*2*t)
x_rand = [np.random.uniform(-1,1) for i in range(len(t))]
x_saw = scisig.sawtooth(2*np.pi*t)
M = len(m)
cors_sin = autocorrelate(x_sin,M)
cors_sqr = autocorrelate(x_sqr,M)
cors_rand = autocorrelate(x_rand,M)
cors_saw = autocorrelate(x_saw,M)

np_cors_sin = np.correlate(x_sin,x_sin,mode='full')
np_cors_sqr = np.correlate(x_sqr,x_sqr,mode='full')
np_cors_rand = np.correlate(x_rand,x_rand,mode='full')
np_cors_saw = np.correlate(x_saw,x_saw,mode='full')

names = ['sin', 'square', 'random', 'sawtooth']
column_names = [' signal',' autocorrelation',' numpy autocorrelation']
y_names = ['value','autocorelation value','autocorelation value']
x_names = ['time','shift','shift']
fig, ax = plt.subplots(3,4,figsize=(20,10), constrained_layout=True)
ax[0,0].plot(t,x_sin)
ax[0,1].plot(t,x_sqr)
ax[0,2].plot(t,x_rand)
ax[0,3].plot(t,x_saw)

ax[1,0].plot(m,cors_sin)
ax[1,1].plot(m,cors_sqr)
ax[1,2].plot(m,cors_rand)
ax[1,3].plot(m,cors_saw)
R = np.arange(-t_len,t_len-1)
ax[2,0].plot(R,np_cors_sin)
ax[2,1].plot(R,np_cors_sqr)
ax[2,2].plot(R,np_cors_rand)
ax[2,3].plot(R,np_cors_saw)
for i in range(3):
  for j in range(4):
    ax[i,j].set_title(names[j]+column_names[i],fontsize = 16,fontweight='heavy')
    ax[i,j].set_xlabel(x_names[i],fontsize = 14,fontweight='light')
    ax[i,j].set_ylabel(y_names[i],fontsize = 14,fontweight='light')
```

```python
f_sin,w_sin = scisig.welch(x_sin,sf)
f_sqr,w_sqr = scisig.welch(x_sqr,sf)
f_rand,w_rand = scisig.welch(x_rand,sf)
f_saw,w_saw = scisig.welch(x_saw,sf)
pnames = [' welch periodogram',' periodogram']

fp_sin,p_sin = scisig.periodogram(x_sin,sf)
fp_sqr,p_sqr = scisig.periodogram(x_sqr,sf)
fp_rand,p_rand = scisig.periodogram(x_rand,sf)
fp_saw,p_saw = scisig.periodogram(x_saw,sf)
fig,ax = plt.subplots(4,2,figsize=(20,10), constrained_layout=True)
ax[0,0].plot(f_sin,w_sin)
ax[1,0].plot(f_sqr,w_sqr)
ax[2,0].plot(f_rand,w_rand)
ax[3,0].plot(f_saw,w_saw)

ax[0,1].plot(fp_sin,p_sin)
ax[1,1].plot(fp_sqr,p_sqr)
ax[2,1].plot(fp_rand,p_rand)
ax[3,1].plot(fp_saw,p_saw)
for i in range(4):
  for j in range(2):
    ax[i,j].set_title(names[i]+pnames[j],fontsize = 16,fontweight='heavy')
    ax[i,j].set_xlabel('frequency',fontsize = 14,fontweight='light')
    ax[i,j].set_ylabel('magnitude',fontsize = 14,fontweight='light')
```
